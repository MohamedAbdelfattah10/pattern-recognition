```python
import speech_recognition as sr
import webbrowser
import librosa.feature
from scipy.io import wavfile
import librosa  
import librosa.display
import matplotlib.pyplot as plt
import matplotlib.style as ms
import pyaudio
import wave

#Recording The Voice 
CHUNK = 1024 #we record 1024 bytes of data
FORMAT = pyaudio.paInt16
CHANNELS = 2
RATE = 44100 #44KBYTES  generated from the Analog Audio signals per second.
RECORD_SECONDS = 5
WAVE_OUTPUT_FILENAME = "output.wav"

p=pyaudio.PyAudio()

stream = p.open(format=FORMAT,channels=CHANNELS,rate=RATE,input=True,frames_per_buffer=CHUNK)
print("* What do you want to look for? *")
frames = []

for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
    frames.append(data)
print("* done *")

wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(p.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()


sampleratee,data=wavfile.read('output.wav')

print(sampleratee)
print(data)

#Conver The Voice TO Text 

r = sr.Recognizer()
sound = sr.AudioFile('output.wav')
#remove nosie 
with sound as source:
    r.adjust_for_ambient_noise(source)
    audio = r.record(source)
    
text=r.recognize_google(audio)
print("YOU SAID: "+r.recognize_google(audio))



#vis orignal signal
ms.use("seaborn-muted")
data,sampling_rate=librosa.load("E:/Abdelfattah/Level 4/Pattern/speech_recognition/output.wav")


mfccs=librosa.feature.mfcc(data,sr=sampling_rate,n_mfcc=40)
plt.figure(figsize=(10,5))
librosa.display.specshow(mfccs,x_axis="time")
plt.colorbar()
plt.title("MFCCS")
plt.tight_layout()
plt.show()



#Search in goole 

try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")
# to search
for j in search(text, tld="co.in", num=1, stop=1, pause=2):
    print(j)
    z=webbrowser.open("https://google.com/search?q=%s" %j)       
    z=webbrowser.open(j)
    
```
